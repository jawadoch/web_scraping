```python
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd
```

```python
def get_page(page):
    url=f"https://wuzzuf.net/a/Experienced-Jobs-in-Egypt?filters%5Bcountry%5D%5B0%5D=Egypt&start={page}"
    page= requests.get(url)
    soup= BeautifulSoup(page.content, 'html.parser')
    list_emp=soup.find_all("div", class_="css-pkv5jc")
    Job_titles=[]
    Companys=[]
    Locations=[]
    Descriptions=[]
    Tags=[]
    for emp in list_emp:
        try:
            Job_titles.append(emp.find('h2', class_='css-m604qf').text)
            Companys.append(emp.find('div', class_='css-d7j1kk').find('a').get_text().replace('-', '').strip())
            Locations.append(emp.find('div', class_='css-d7j1kk').find('span').get_text())
            Descriptions.append(emp.find('div', class_='css-y4udm8').find_all('div')[1].text)
            liste=[]
            for i in range(len(emp.find('div', class_='css-y4udm8').find_all(class_='css-n2jc4m'))):
                liste.append(emp.find('div', class_='css-y4udm8').find_all(class_='css-n2jc4m')[i].text)
            if emp.find('div', class_='css-y4udm8').find(class_='css-o1vzmt eoyjyou0'):
                liste.append(emp.find('div', class_='css-y4udm8').find(class_='css-o1vzmt eoyjyou0').text)
            Tags.append(liste)
        except Exception as e:
            print(print(f"An unexpected error occurred: {e}"))
    Jobs={"Job_title":Job_titles, "Company":Companys, "Location":Locations, "Description": Descriptions, "Tag":Tags}
    return pd.DataFrame(Jobs)
```

```python
df = [get_page(i) for i in range(1,10)]
df=pd.concat(df)
df.index = range(1, len(df) + 1)
df
```

```python
dff = df.explode('Tag')
# Réinitialiser l'index si nécessaire
dff = dff.reset_index(drop=True)
dff
```

```python
def extract_location(location_str):
    elements = location_str.split(', ')
    if len(elements) == 3:
        return elements
    elif len(elements) == 2:
        return [None] + elements   # Ajouter None si la rue est manquante

```

```python
# Appliquer la fonction à la colonne 'Location'
dff[['Street', 'City', 'Country']] = pd.DataFrame(dff['Location'].apply(extract_location).tolist(), index=dff.index)

# Supprimer la colonne 'Location' d'origine
dff = dff.drop('Location', axis=1)


```
