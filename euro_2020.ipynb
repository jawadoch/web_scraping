```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url="https://fr.wikipedia.org/wiki/Championnat_d%27Europe_de_football_2020"
page= requests.get(url)
content = page.text
soup = BeautifulSoup(content, 'lxml')
```

```python
divv = soup.find('div', class_='mw-content-ltr mw-parser-output')
```

```python
matches=divv.find_all('table', class_=lambda x: x is None)
len(matches)
```

```python
matches4=[matches[i] for i in range(4, 40)]
```

```python
home_team=[]
score=[]
away_team=[]
for match in matches4:
    try:
        home_team.append(match.find_all('td')[1].find('a').text.strip())
        score.append(match.find_all('td')[2].find('b').text.strip())
        
        away_team.append(match.find_all('td')[3].find_all('a')[1].text.strip())
    except:
        pass
matches1={'home_team':home_team, 'score':score, 'away_team':away_team}
df=pd.DataFrame(matches1)

    
```

```python
df
```

```python
matches5=[matches[i] for i in range(41, 56)]
```

```python
home_team1=[]
score1=[]
away_team1=[]
for match in matches5:
    try:
        home_team1.append(match.find_all('td')[1].find('a').text.strip())
        score1.append(match.find_all('td')[2].find('b').text.strip().replace('a. p.',''))
        away_team1.append(match.find_all('td')[3].find_all('a')[1].text.strip())
    except:
        pass
matches2={'home_team':home_team1, 'score':score1, 'away_team':away_team1}
df2=pd.DataFrame(matches2)

    
```

```python
df2
```

```python
result_df = pd.concat([df, df2], ignore_index=True)
result_df
```
